### Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
```

```python
movies.head()
```

```python
ratings.head()
```

```python
best_movie = ratings[ratings.rating == 5.0 ]
joined_data = movies.merge(best_movie, on='movieId', how='left')

print('Лучший фильм:', joined_data['title'].value_counts().index[0])
```

### Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power_data = pd.read_csv('power.csv')
power_data.head()
```

```python
countries_final = power_data[ ((power_data['country']=='Latvia') | (power_data['country']=='Lithuania') | (power_data['country']=='Estonia')) ]
categories_final = countries_final[ ((countries_final['category']==4) | (countries_final['category']==12)) | (countries_final['category']==21) ]
year_final = categories_final[ (categories_final['year'] > 2004) & (categories_final['year'] < 2011) ]
quantity_final = year_final[ (year_final['quantity'] >= 0) ]
sum_by_country = year_final.groupby(['country']).sum()
sum_by_country = sum_by_country.filter( items = ['country', 'quantity'])

print('Суммарное потребление по всем странам:', format(sum_by_country['quantity'].sum()))
```
